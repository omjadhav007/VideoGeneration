```python
from TTS.api import TTS
import os
```

```python
def generate_voice(script, output_audio_path):
    # Using Tacotron 2 for TTS
    tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC")
    alignment_data = tts.tts(text=script, file_path=output_audio_path, speaker_wav=None, return_raw_alignments=True)
    return alignment_data
```
